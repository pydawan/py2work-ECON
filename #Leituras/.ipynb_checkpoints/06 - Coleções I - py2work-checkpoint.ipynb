## Listas

 ### Fundamentos de Lista
 
 Imagine a lista como um gaveteiro magico onde você pode guardar outras estruturas de dados até mesmo outras listas!<br>
  Para criar uma lista basta utilizar sua sintaxe, lista = [args*], após isso teremos com ela em mãos a primeira coisa que você pode fazer seria acessar seus elementos esse conceito é o de indice,posições do nosso gaveteiro magico.<br>

 O ultimo assunto que gostaria de abordar sobre fundamento de listas é **objetos e referencia** pois em geral há um grande problema o qual irá ficar mais claro abaixo entretanto. Vale dizer a função **is** tem como **objetivo verificar** se os objetos referenciados estão **alocados no mesmo espaço de memoria**.<br>
 Basicamente quando declaramos variaveis o python reserva um espaço da memoria do seu computador para aquela entidade ao declarar que a = b estamos colocando as variaveis no mesmo espaço de memoria e ao fazer isso uma alteração em a leva uma alteração em b isso pode causar erros extremamentes difíceis de se identificar logo deve ter atenção.<br>
 Adjunto a isso em listas entre outros objetos há um problema quando você criar listas iguais o python irá otimizar automaticamente seu código colocando elas no mesmo espaço de memoria. ~~leiaoartigodoimeamigao!~~

 ### Métodos em lista
 Acredito que os principais métodos em listas são **list,append,extend,sort,reverse,index,remove e pop** irei utilizar e explicar sua funcionalidade ao código abaixo.   ~~len é muito util!~~

```python
"""
    Lista
     -> Fundamentos
     -> Metodos
"""

##
### Fundamentos
##

"""
    Fundamentos
     -> Criando
     -> Acessando posições
     -> Alocação de memoria
"""
##Criando uma lista
primeira_lista = []

lista_nomes = ["BTG" , "Itau", "Goldman", "Morgan", "Rothschild"]
lista_nested = ["Eu trabalho no :", lista_nomes]

## Os argumentos que estão dentro da lista são modos de acessar elementos da lista!
lista_acessos = [ lista_nomes[0] ,lista_nomes[1:4] ,lista_nomes[:3] ,lista_nomes[-1] ,lista_nested[1][2]]


#Vamos utilizar um
for i in lista_acessos:
    print("Vamos imprimir : {}".format(i))
    

print("\n")

## Objetos e referencia!

nome1 = "loop"
nome2 = "loop"

print("Os nome1 e nome2 tem referencial {} , e seu objeto e tem uma estrutura {}".format(nome1 is nome2,nome1==nome2) , "\n")



lista1 = [1,2,3]
lista2 = [1,2,3]
print("A lista 1 e 2 tem referencial {} , e seu objeto e tem uma estrutura {}".format(lista1 is lista2,lista1==lista2),"\n")


print(lista1 is lista2)

lista3 = [4,5]
lista4 = lista3

print("A lista 3 e 4 tem referencial {} , e seu objeto e tem uma estrutura {}".format(lista3 is lista4,lista3==lista4),"\n")


#Clonando lista
lista5 = lista3[:]
print("A lista 5 e 3 tem referencial {} , e seu objeto e tem uma estrutura {}".format(lista5 is lista3,lista5==lista3),"\n")

"""
    #******#
"""
print("\n")
##
### Metodos
##

"""
    Metodos
     -> append : Adiciona um item ao fim da lista
     -> extend : Adiciona uma lista de itens no fim da lista
     -> sort : Ordena as lista alfa-numerica
     -> reverse : Inverte a lista
     -> index : Retorna a posição de um elemento
     -> remove : Remove um item da lista
     -> pop : Mostra o ultimo item da lista e o retira
     -> insert : Injeta em uma posição especifica um elemento
"""

##Utilizando os metodos

ipca = [9,10,25,8,12,1,3,7,9,-1,11,4]
ano = list(range(2000,2012)) #A função list converte outras estruturas em lista! como esse objeto

#append
print(lista_nomes)
lista_nomes.append('Santander')
print(lista_nomes)

#extend
lista_teste = []
lista_teste.extend(lista_nomes)
print(lista_nomes)

#sort
lista_nomes.sort()

#reverse
lista_teste.reverse()

#index
print(lista_nomes.index("Itau"))

#remove
lista_nomes.remove("Santander")

#pop
print(lista_teste.pop())

#insert
lista_teste.insert(2,"BB")
```

## Tuplas

### Fundamentos de tuplas

  <div class="alert alert-block alert-danger">
<b>Tuplas:</b> Tem a mesma definição de similar a lista entretanto elas são objetos imutaveis!
</div>

### Métodos em tuplas
 Não há métodos para tuplas entretanto há uma função de conversão chamada **tuple** muito usada além disso há um certo trick para modificação de tuplas o que fazemos para modificar tuplas é utilizar nossa referencia sobreescrevendo o objeto!


```python
"""
    Tuplas
     -> Tuplas
     -> Zip
"""

##
### Tuplas
##

"""
    Tuplas
     -> Criando
     -> Acessando posições
"""

#<>Criando tuplas
primeira_tupla =("MaNinja",999,"Equity",["envelope",999,"Markov","Marshal"])
lista_titulos = tuple(["LFT","NTN"]) #Função de conversão

print(type(primeira_tupla))

#Olhe que engraçado
string = ("sou_uma_string")
tupla = ("sou_uma_string",)

print(type(string), type(string))  #função type checa o tipo de dado


#<> Acessando posicao : similar a lista
print(primeira_tupla[1])
print(primeira_tupla[3][0])



#Imutavel
####  primeira_tupla[1] = 10 retire os comentarios e teste!


#Update de tuplas

#   antiga
nomes = ("rafa","greg")
print(nomes)
#   Update
nomes = ("rafael","legal","SQN")
print(nomes)


"""

"""
print("\n"*2,"Começando Zip")
##
### Zip
##
"""
zip é uma função integrada que recebe duas ou mais sequências e devolve uma lista de tuplas
 onde cada tupla  contém um elemento de cada sequência. 
O nome da função tem a ver com o zíper, que se junta e encaixa duas carreiras de dentes. -Pense em Python
"""

#Usando função zip
lista_fundos = ["Goldman","Itau","BTG"]
ativos = ["Renda fixa","Derivativos" , "NTN-B"]

objeto_zip = zip(lista_fundos,ativos)
tupla_zipada = tuple(objeto_zip)    #poderia ser uma lista

print(objeto_zip , tupla_zipada)


#FOR EM TUPLAS
for i in tupla_zipada:
    print(i)
    
for i,j in tupla_zipada:
    print("X : {} , Y: {}".format(i,j))
```

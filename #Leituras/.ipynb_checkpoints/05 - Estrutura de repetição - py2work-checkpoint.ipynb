 Vamos **separar a situação em duas partes** bases para simplificar seu entendimento ~~diminuirmeutrampotbmbruxo~~
 
 1. Sabemos que temos que entrar na quinta rua enquanto não chegar nela não podemos entrar.
 2. Dada uma lista de anos a empresa deve pagar uma quantia X nestes anos pares!.

```python
"""
Sintaxe While

while (expressão):
    #CODE
else:
    #CODE

Esse else no final é opcional caso queiramos que algum código seja executado apos o termino de nosso loop!

Vale ressaltar que podemos combinar muitos whiles um dentro de outro além de juntar com outras
  estruturas mas acredito que você vai descobrir isso ao longo do curso ou de links extras!
"""

## Parte 1

conta_rua = 0


while conta_rua <= 4:
    conta_rua += 1
    print(conta_rua)
else:
    print("Finalizamos o código")


"""
Sintaxe for

for var in (estrutura_de_dado):
    #CODE
"""

#Antes de ir para o problema quero que você rode esses codigos e compreenda o mecanismo

## Rode essa parte (A)

ipca = [1,2,3,4]
juros = [999,000]

for i in ipca:
    print(i)

## Rode agora essa parte (B)

for i in ipca:
    for j in juros:
        print(i,j)

## Parte 2 ; Agora sim é o problema principal

anos = [2011,2012,2013,2014,2015,2016,2017,2018,2019]

for i in anos:
    if i%2 ==0:
        print("Temos que pagar em {}".format(i))

## Parte 2.1 ; Vamos além!

caixa = 100

for i in anos:
    if i%2 ==0:
        caixa = caixa - 10
        print("Temos que pagar em {} e nosso caixa atual é {:0.2f}".format(i,caixa))


## Parte 2.2 ; Vamos subir mais ainda bruxo!

aplicacao = 0.12

for i in anos:
    if i%2 ==0:
        caixa = caixa * (1+ aplicacao)
        caixa = caixa - 13
        print("Temos que pagar em {} e nosso caixa atual é {:0.2f}".format(i,caixa))
```

## Analisando o código!



**Parte 1**: Inicializamos uma variavel contador a qual executa um while ou seja o bloco dentro até a condição não ser mais verdadeira por fim finalizamos com else que é opcional.


Antes de iniciar rodar nosso problema principal compreendemos o comportamento do for

**Parte 2**: Basicamente o que o código faz é entrar no indice da lista isso é transformar o i naquele numero então obtemos o resto da divisão do i para obter se o numero é par caso for par iremos aplicar o código que é uma impressão de tela.

Na parte 2.1 aplicamos uma variavel caixa de onde a empresa paga sua divida ,no caso é 10 , enquanto na parte 2.2 fizemos algo melhor ainda colocamos que o caixa da empresa consegue rentabilizar a uma aplicação.
